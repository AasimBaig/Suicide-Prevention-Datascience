```python
color = ['blue','green']
labels = ['below 14', '14-25', '25-37', '37-48', '48-60','60 above']
men_means = [899,4682,27343,30659,18439,7107]
women_means = [821,4548,17527,11723,5248,2654]


x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(12,10))
rects1 = ax.bar(x , men_means, width, label='Men',color='#6c0e06',edgecolor ="w")
rects2 = ax.bar(x + width, women_means, width, label='Women',color='#044247',edgecolor ="w")

ax.legend()
ax.set_xticklabels(('','below 14', '14-25', '25-37', '37-48', '48-60','60 above'))
ax.set_xlabel("Age")
ax.set_ylabel("Count")



def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)

ax.set_title("Age Demographic of Suicide")
fig.tight_layout()

plt.show()

fig.savefig("charts/suicide_demographics.jpg")
```
