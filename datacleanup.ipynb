```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cleandata = pd.read_csv("data-cleaned/cleandata2.csv")
statedata = pd.read_csv("data-cleaned/data_clean-cleaned-states.csv")
total_crimehead = pd.read_csv("data-cleaned/total_crimeheads.csv")
state_totatl_crimehead = pd.read_csv("data-cleaned/total_crimeheads-clean-cleaned.csv")
```

```python
cleandata.head()
```

```python
df = cleandata[cleandata["States/UTs"]=="Total (All India)"]
df = df.sort_values(by="Grand Total",ascending=False)
df.head()
```

```python
plt.figure(figsize=(10,10))

sns.barplot(df.CrimeHead,df["Grand Total"],palette="jet_r")
plt.title("Nature of suicides",fontsize=20)

plt.xlabel("Types")
plt.xticks(rotation=88)

plt.plot()
```
